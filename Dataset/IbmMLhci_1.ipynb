for label

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset['State'] = le.fit_transform(dataset['State'])   # for converting in label
```

```python
dataset.head()
```

for onehot

```python
x=dataset.iloc[:,0:4].values
y=dataset.iloc[:,4:5].values
```

```python
x.dtype
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(x[:,3:4]).toarray()
x =np.delete(x,3,axis=1)
x =np.concatenate((z,x),axis=1)
```

```python
x   # converting in to array
```

Train, Test & Split

```python
from sklearn.model_selection import train_test_split   # preprocessing  split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2, random_state=0)  #random_state is for soft linking

#hyper parameter change
```

Model Building

IMporot libraries

add layers

```python
model.add(Dense(units=4,activation='relu'))   # model is name is artif node connect to dense unit is variable relu is use to activation fun  input layer
```

```python
model.add(Dense(units=12,activation='relu'))   # 1st hidden layer
```

```python
model.add(Dense(units=8,activation='relu'))      # 2nd hidden layer
```

```python
model.add(Dense(units=9,activation='relu'))     # 3rd hidden layer
```

output Layer

```python
model.add(Dense(units=1))
```

Compile The Model

```python
model.compile(optimizer = 'adam' , loss = 'mse', metrics = ['mse'])   # adam is optimizer mean square error adam is same like gradient codt funtion is loss
#mse mean square error
```

Fit The model

```python
model.fit(x_train, y_train, batch_size = 15, epochs = 100)   # epochs  one pass to whole dataset
```

```python
model.fit(x_train,y_train, batch_size=15,epochs=800)   # at last give 500 epochs or 550
```

search good value of mse

Test the model

```python
 y_pred=model.predict(x_test)  # test our model on testing data
```

```python
 y_pred   # predicted value
```

```python
  y_test   # original value
```

Calculate R2_score or cofficient of determination

```python
from sklearn.metrics import r2_score # show diff in original and predic value
accuracy = r2_score(y_test,y_pred)
```
