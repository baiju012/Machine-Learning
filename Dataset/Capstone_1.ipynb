```python
import pandas as pd
import numpy as np
import sklearn
import math as m
```

```python
df=pd.read_csv("train.csv")
df=df.rename(columns={'Offer %': 'offer'})
df.head(n=2)
```

```python
df.columns
```

```python
for i in df.columns:
  print(i,df[i].isna().sum())
```

```python
df2=df.dropna(subset=['maincateg'])
```

```python
df2.head(n=1)
```

```python
df2=df2[df2.isnull().sum(axis=1)<2]
df3=df2.replace('%','',regex=True)
```

```python
df3.dropna(inplace=True)
```

```python
df3.head(n=1)
```

```python
X_col=['Rating','actprice1','offer','norating1','noreviews1','star_5f','star_4f','star_3f','star_2f','star_1f']
#y_col=['price1']
x=df3[X_col]
y=df3.price1
print(x.shape)
print(y.shape)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.35,random_state=1)
```

## RIDGE REGRESSION

```python
# evaluate an ridge regression model on the dataset
from numpy import mean
from numpy import std
from numpy import absolute
from pandas import read_csv
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import Ridge
from sklearn.linear_model import LogisticRegression
```

```python
model = Ridge(alpha=1.0)
cv = RepeatedKFold(n_splits=2, n_repeats=1, random_state=1)
scores = cross_val_score(model, X_train, Y_train, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
scores = absolute(scores)
mean(scores)
```

## LOGISTIC REGRESSION 

```python
regressor= LogisticRegression()
regressor.fit(X_train,Y_train)
```

##LINEAR REGRESSION

```python
reg=linear_model.LinearRegression()
reg.fit(X_train,Y_train)
```

```python
reg.score(X_test,Y_test)
y_pred=reg.predict(X_test)
```

```python
mse = sklearn.metrics.mean_squared_error(Y_test,y_pred)
rmse = m.sqrt(mse)
print(rmse)
```
